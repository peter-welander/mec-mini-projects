# Iris Flower Dataset

Toy Dataset: Iris Dataset:
https://en.wikipedia.org/wiki/Iris_flower_data_set

Objective: Classify a flower as belonging to one of the three classes.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("iris.csv")
```

```python
# (Q) How many data points and features?

print (iris.shape)
```

```python
# Q: What are the columns?

print (iris.columns)
```

```python
# (Q) How many data points for each class are present?
# (or) How many flowers for each species are present?

iris["Species"].value_counts()

# balanced-dataset vs imbalanced
# Iris is a balanced dataset as the number of data points for every class is the same.
```

# (3.2) 2-D Scatter Plot


```python
# 2-d scatter plot
# ALWAYS understand the axis: labels and scale.

iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
plt.show()

# cannot make much sense out it.
# what if we color the points by their class-label flower type?
```

```python
# 2-d scatter plot with color-coding for each flower type/class.
# Here 'sns' corresponds to seaborn.
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue='Species') \
    .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
    .add_legend()
plt.show()

# Blue and green can be easily separated by drawing a line
# Can we draw multiple 2-d scatter plots for each combination of features?
# How many combinations exist? C(4, 2) = 6
```

## Observation:

1. Using Sepal length and width we can distinguish setosa flowers from others.
1. Separating versicolor from virginica is harder since they have a lot of overlap
